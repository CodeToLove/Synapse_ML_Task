## Task 2.2 - Operations

Write a code to give me the number of rows that have the release_year as 2021 and type as Movie.

```python
# Your output will be 146, if correct
filtered_rows = df[(df['release_year'] == 2021) & (df['type'] == 'Movie')]

# Get the number of rows that match the conditions
num_rows = filtered_rows.shape[0]

print("Number of rows with release year as '2021' and type as 'Movie':", num_rows)
```

Write a code to create a list of all the movies that have the word 'Avengers' in it.

```python
# Your output will contain 2 movies, if correct

avengers_movies = df[df['title'].str.contains('Avengers')]

# Extract the 'title' column as a list
avengers_movie_list = avengers_movies['title'].tolist()

print("Movies with 'Avengers' in the title:")
for movie in avengers_movie_list:
    print(movie)
```

Display a new dataframe, that contains only those movies that are present in the Top 5 countries.

(Don't forget to reset the indices again)

```python
grouped_data = df['country'].value_counts()



# Create a new DataFrame containing the top 5 countries
top_countries = grouped_data.nlargest(5).index.tolist()
top_country_data = grouped_data.loc[top_countries]
topCountries = pd.DataFrame(top_country_data)
topCountries
```

## Task 2.3 - Chart

Use the new dataframe to create a Pie Chart on the basis of number of Movies/TV Shows per country.

```python
grouped_data = df['country'].value_counts()



# Create a new DataFrame containing the top 5 countries
top_countries = grouped_data.nlargest(5).index.tolist()
top_country_data = grouped_data.loc[top_countries]

# Plot a pie chart
plt.figure(figsize=(10, 6))
plt.pie(top_country_data, labels=top_country_data.index, autopct='%1.1f%%')
plt.title("Distribution of Movies/TV Shows per Country (Top 5)")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
